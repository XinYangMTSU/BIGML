## Load fmri dataset from ABIDE Data Repository

```python
import warnings
warnings.filterwarnings("ignore")

!pip install nilearn

from nilearn.datasets import fetch_abide_pcp
# Fetch the full data and update phenotypic data and cross_validation
abide = fetch_abide_pcp(derivatives = ['func_preproc'], pipeline = 'cpac', quality_checked = True, n_subjects = 120)

y = abide.phenotypic['DX_GROUP']
```

```python
import pandas as pd
import numpy as np
names = ['DX_GROUP','func_preproc']
data = pd.DataFrame([y,abide.func_preproc]).transpose()
data.columns = names
ASD =  data[data['DX_GROUP']==1]['func_preproc'] #403
TD = data[data['DX_GROUP']==2]['func_preproc'] #468
```

## Apply CanICA on the ASD data

```python
from nilearn.decomposition import CanICA

canica_asd = CanICA(n_components=20,
                memory="nilearn_cache", memory_level=2,
                verbose=10,
                mask_strategy='template',
                random_state=0)

#canica.fit(abide.func_preproc)
canica_asd.fit(ASD)

# Retrieve the independent components in brain space. Directly
# accesible through attribute `components_img_`.
canica_components_img_asd = canica_asd.components_img_
canica_components_img_asd.to_filename('canica_resting_state.nii.gz')
```

```python
# Each region index is stored in index_
from nilearn.regions import RegionExtractor

extractor_asd = RegionExtractor(canica_components_img_asd, threshold=0.5,
                            thresholding_strategy='ratio_n_voxels',
                            extractor='local_regions',
                            standardize=True, min_region_size=1350)
# Just call fit() to process for regions extraction
extractor_asd.fit()
# Extracted regions are stored in regions_img_
regions_extracted_img_asd = extractor_asd.regions_img_
# Each region index is stored in index_
regions_index_asd = extractor_asd.index_
# Total number of regions extracted
n_regions_extracted_asd = regions_extracted_img_asd.shape[-1]
```

```python
regions_index_asd
```

```python
# Region Extraction with Group ICA maps
from nilearn.connectome import ConnectivityMeasure

time_series_ICA = []

for i in range(len(abide.func_preproc)):
    print("i: " + str(i))
    # call transform from RegionExtractor object to extract timeseries signals
    subject = extractor_asd.transform(abide.func_preproc[i])
    # saving each subject correlation to correlations
    time_series_ICA.append(subject)
```

## Estimate explained variance per component and plot using matplotlib

```python
scores = canica_asd.score(abide.func_preproc, per_component=True)

# Plot the scores
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter

plt.figure(figsize=(4, 4))
positions = np.arange(len(scores))

plt.barh(positions, scores)
plt.ylabel('Component #', size=12)
plt.xlabel('Explained variance', size=12)
plt.yticks(np.arange(20))
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.3f'))
plt.tight_layout()

show()
```

## Diction Learning

```python
from nilearn.decomposition import DictLearning

dict_learning_asd = DictLearning(n_components=20,
                             memory="nilearn_cache", memory_level=2,
                             verbose=1,
                             random_state=0,
                             n_epochs=1,
                             mask_strategy='template')

print('[Example] Fitting dicitonary learning model')
#dict_learning.fit(abide.func_preproc)
dict_learning_asd.fit(ASD)

print('[Example] Saving results')
# Grab extracted components umasked back to Nifti image.
# Note: For older versions, less than 0.4.1. components_img_
# is not implemented. See Note section above for details.
dictlearning_components_img_asd = dict_learning_asd.components_img_
dictlearning_components_img_asd.to_filename('dictionary_learning_resting_state.nii.gz')
```

```python
# Each region index is stored in index_
from nilearn.regions import RegionExtractor

extractor_dict_asd = RegionExtractor(dictlearning_components_img_asd, threshold=0.5,
                            thresholding_strategy='ratio_n_voxels',
                            extractor='local_regions',
                            standardize=True, min_region_size=1350)
# Just call fit() to process for regions extraction
extractor_dict_asd.fit()
# Extracted regions are stored in regions_img_
regions_extracted_img_dict_asd = extractor_dict_asd.regions_img_
# Each region index is stored in index_
regions_index_dict_asd = extractor_dict_asd.index_
# Total number of regions extracted
n_regions_extracted_dict_asd = regions_extracted_img_dict_asd.shape[-1]
```

```python
from nilearn.connectome import ConnectivityMeasure

time_series_Dict = []

for i in range(len(abide.func_preproc)):
    print("i: " + str(i))
    # call transform from RegionExtractor object to extract timeseries signals
    subject = extractor_dict_asd.transform(abide.func_preproc[i])
    # saving each subject correlation to correlations
    time_series_Dict.append(subject)
```
